# Download Datasets Notebook

Run this notebook **once** to download and save all datasets locally.

- Sentiment: https://huggingface.co/datasets/tweet_eval
- Emotion: https://huggingface.co/datasets/dair-ai/emotion  
- GoEmotions: https://huggingface.co/datasets/go_emotions

> Raw files are saved to `data/raw/` and excluded from git.

### Importing the datasets 

```python
import os 
from datasets import load_dataset
```

### Creating the `data/raw/` folder for the dataset

```python
os.makedirs('../data/raw', exist_ok=True)
```

### Downloading the datasets from huggingface

```python
print('Downloading Sentiment Dataset...')
sentiment_ds = load_dataset("cardiffnlp/tweet_eval", "sentiment")
for split in ["train", "validation", "test"]:
    sentiment_ds[split].to_pandas().to_csv(f"../data/raw/sentiment_{split}.csv", index=False)

print('Sentiment Saved!')
```

```python
print('Downloading Emotion Dataset...')
emotion_ds = load_dataset('dair-ai/emotion')
for split in ["train", "validation", "test"]:
    emotion_ds[split].to_pandas().to_csv(f"../data/raw/emotion_{split}.csv", index=False)

print('Emotion Saved!')   
```

```python
print('Downloading GoEmotions Dataset...')
go_ds = load_dataset("google-research-datasets/go_emotions", "simplified")
for split in ["train", "validation", "test"]:
    go_ds[split].to_pandas().to_csv(f"../data/raw/go_emotions_{split}.csv", index=False)
    
print('GoEmotions Saved!')
```
