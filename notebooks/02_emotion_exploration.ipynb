# Emotion Dataset Exploration
Dataset: dair-ai/emotion<br>
Source: https://huggingface.co/datasets/dair-ai/emotion

### Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the datasets (All Splits)

```python
df_train = pd.read_csv('../data/raw/emotion_train.csv')
df_val = pd.read_csv('../data/raw/emotion_validation.csv')
df_test = pd.read_csv('../data/raw/emotion_test.csv')

print('Train Size: ', df_train.shape)
print('Validation Size: ', df_val.shape)
print('Test Size: ', df_test.shape)
```

### Preview of Datasets

```python
label_map = {0: 'Sadness', 1: 'Joy', 2: 'Love', 3: 'Anger', 4: 'Fear', 5: 'Surprise'}
df_train['label_name'] = df_train['label'].map(label_map)

df_train.head()
```

### Class distribution Graph

```python
df_train['label_name'].value_counts().plot(kind='bar', color=['steelblue', 'gold', 'hotpink', 'tomato', 'slategray', 'mediumpurple'])
plt.title('Emotion Class Distribution (Train)')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print(df_train['label_name'].value_counts())
```

### Observations Noted

Observations
- Class distribution: Joy > Sadness > Anger > Fear > Love > Surprise
- Joy and Sadness are the dominant emotion classes in the dataset.
- Love and Surprise are significantly underrepresented compared to other emotions. <br>

Mitigation strategies:
- This class imbalance may bias the model toward majority emotion classes
- Use `class_weight='balanced'` during training
- Monitor macro F1-score instead of relying only on accuracy
- Consider oversampling or data augmentation for minority classes such as Love and Surprise
